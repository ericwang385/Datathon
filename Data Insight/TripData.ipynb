Hello!  This is the code to open one month of the NYC Taxi Data for one year (in our case, 2013).  For our model, we only needed some columns of data, so the first few cells here simply drop the columns we don't need, and save the resulting slimmed-down csv file with a new name.

```python
import pandas as pd

df = pd.read_csv('trip_data_file.csv', sep = ',' , header = 0) #open the file
df.head() #have a look at the first few rows
```

Now we can see all the columns, it's easy to identify the ones we need to drop.  That's what happens next, and then we save the new file.

```python
df.drop(['medallion',' hack_license', ' vendor_id', ' rate_code', ' store_and_fwd_flag', ' trip_distance'], axis = 1, inplace = True) #
df.head()
#'medallion','hack_license', 'vendor_id', 'rate_code', 'store_and_fwd_flag', 'trip_distance'
```

```python
df.dtypes
```

```python
df.to_csv('TripDataFields.csv', index = False)
```

What follows is some extra lines of code that I used to extract a 1% sample of January's data.  First of all, I shuffled the rows and then extracted the sample, saving it as a new csv file.

```python
df = pd.read_csv('TripDataFields.csv', sep = ',' , header = 0)
df.head()
```

```python
df.tail()
```

```python
from sklearn.utils import shuffle  #shuffle rows
df = shuffle(df)
```

```python
df.head()
```

```python
train = df.sample(frac = 0.01, random_state = 42)  #select a training sample of 25%
```

```python
train.shape
```

```python
train.to_csv('TripDataSample.csv', sep = ',')
```
